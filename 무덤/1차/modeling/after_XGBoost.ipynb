```python
after = pd.read_csv('../data_merge/after_data.csv')
after_data = after.drop(['Date','ETF'], axis=1)
after_target = after['ETF']
len_ = len(after_data)
x_train, x_test, y_train, y_test = after_data[:round((len_*0.8))], after_data[round((len_*0.8)):], after_target[:round((len_*0.8))], after_target[round((len_*0.8)):]


xgb = XGBRegressor(n_estimators=400, max_depth=3, learning_rate=0.1)
xgb.fit(x_train, y_train)
pred = xgb.predict(x_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)
print(f'mse : {mse} \nrmse : {rmse} \nr2 : {r2}')
```

```python
after = pd.read_csv('../data_merge/after_data.csv')
after_data = after.drop(['Date','ETF'], axis=1)
after_target = after['ETF']
x_train, x_test, y_train, y_test = train_test_split(after_data, after_target, random_state=42, test_size=0.2)

xgb = XGBRegressor(n_estimators=400, max_depth=3, learning_rate=0.1)
xgb.fit(x_train, y_train)
pred = xgb.predict(x_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)
print(f'mse : {mse} \nrmse : {rmse} \nr2 : {r2}')
```
