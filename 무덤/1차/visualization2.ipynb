```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

###  before data 시각화

```python
before = pd.read_csv('./data_merge/before_data.csv')
before_data = before.drop('ETF', axis=1)
before_target = before['ETF']
before
```

```python
#히트맵
before_withoutDate = before.drop(['Date'], axis=1)
corr_before = before_withoutDate.corr(method='pearson')
sns.heatmap(corr_before,annot=True, cmap='Blues')
plt.savefig("heatmap")
```

```python
a = corr_before.style.background_gradient(cmap='Blues')
a
```
